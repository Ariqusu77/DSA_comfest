```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load the GeoDataFrames
gdf1 = gpd.read_file('geojson-indonesia-master\geojson-indonesia-master\city-regency-simplified\id-jk-jakbar.min.geojson')
gdf2 = gpd.read_file('geojson-indonesia-master\geojson-indonesia-master\city-regency-simplified\id-jk-jakpus.min.geojson')
gdf3 = gpd.read_file('geojson-indonesia-master\geojson-indonesia-master\city-regency-simplified\id-jk-jaksel.min.geojson')
gdf4 = gpd.read_file('geojson-indonesia-master\geojson-indonesia-master\city-regency-simplified\id-jk-jaktim.min.geojson')
gdf5 = gpd.read_file('geojson-indonesia-master\geojson-indonesia-master\city-regency-simplified\id-jk-jakut.min.geojson')

# Merge the GeoDataFrames
Jakarta_gdf = gpd.GeoDataFrame(pd.concat([gdf1, gdf2, gdf3, gdf4, gdf5], ignore_index=True), crs=gdf1.crs)

# Print the merged GeoDataFrame
print(Jakarta_gdf)
```
